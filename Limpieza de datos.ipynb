# Limpieza de Datos

```python
import pandas as pd
```

```python
df_FIFA = pd.read_csv("data/mundiales.csv")
df_FIFA
```

#### Checar NULL

```python
df_FIFA[df_FIFA["Local"].isnull()]
```

```python
# Checar tipos de datos
df_FIFA.dtypes
```

```python
#Quitar los espacios en blanco de los nombres
df_FIFA["Local"] = df_FIFA["Local"].str.strip()
df_FIFA["Visitante"] = df_FIFA["Visitante"].str.strip()
```

```python
### Arreglar marcador
df_FIFA["Marcador"] = df_FIFA["Marcador"].str[:4]
df_FIFA
```

```python
df_FIFA[["GolesLocal","GolesVisita"]] = df_FIFA["Marcador"].str.split(":",expand=True)
df_FIFA.drop(columns=["Marcador"],inplace=True)
df_FIFA
```

```python
#Cambiar el tipo de datos de los goles a enteros
df_FIFA = df_FIFA.astype({"GolesLocal":int, "GolesVisita":int})
df_FIFA
```

```python
df_FIFA.dtypes
```

```python
df_FIFA["GolesTotales"] = df_FIFA["GolesLocal"]+df_FIFA["GolesVisita"]
df_FIFA
```

```python
df_FIFA.to_csv("data/FIFAWC.csv",index=False)
```
